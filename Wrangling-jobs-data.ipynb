## Clean
### Issue 1: Start date
#### Define
- Extract a list of start date values that need to be replaced by 'ASAP'
- Run a for loop over all the extracted values replaceing them. 

#### Code

```python
# Checking initial values of start date
df.StartDate.value_counts()[-20:]
```

```python
# Creating a copy of the original df, to preserve original file
df_clean = df.copy()

# Create a list of values in start date that are synonimous to ASAP
asap_values = list(df_clean[df_clean.StartDate.fillna('').str.strip().str.lower().str.contains('immed') | \
df_clean.StartDate.fillna('').str.strip().str.lower().str.contains('soon') | \
df_clean.StartDate.fillna('').str.strip().str.lower().str.contains('upon hiring') | \
         df_clean.StartDate.fillna('').str.strip().str.lower().str.contains('open') | \
df_clean.StartDate.fillna('').str.strip().str.lower().str.contains('asap') ].StartDate.unique())#.StartDate.fillna('').unique()
```

```python
len(asap_values) # number of values that need to be replaced
```

```python
# loop to replace all values in the df_clean dataset.
for value in asap_values:
    df_clean.StartDate.replace(value,'ASAP',inplace=True)
```

#### Test

```python
df_clean.StartDate.value_counts().plot(kind='pie');
```

### Issue 2: Rename non-descriptive column names
#### Define
- Convert field names to descriptive ones. JobRequirment,ApplicationP, AboutC,RequiredQual
- convert: JobRequirment-JobRequirement, ApplicationP-ApplicationProcess, RequiredQual-RequiredQualification, AboutC-AboutCompany
#### Code

```python
replace_cols = { 'JobRequirment':'JobRequirement',
               'ApplicationP':'ApplicationProcess',
               'AboutC':'AboutCompany',
               'RequiredQual':'RequiredQualification'}
df_clean.rename(replace_cols,axis=1,inplace=True)
```
